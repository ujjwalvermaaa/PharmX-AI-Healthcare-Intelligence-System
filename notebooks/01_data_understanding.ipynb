## Data Understanding and Initial Data Audit

This notebook performs the initial exploration and inspection of all datasets used in the Healthcare AI Analytics Platform project.

The objective of this step is to understand the structure, quality, and relationships of the data before performing data cleaning, feature engineering, and machine learning.

The following aspects are examined in this notebook:

• Dataset sizes and shapes  
• Column data types  
• Missing values  
• Duplicate records  
• Unique categorical values  
• Statistical summaries of numerical variables  
• Dataset relationships and key linkages

This step ensures that the data is well-understood and ready for the next stage of preprocessing and cleaning.

---

## Datasets Used in This Project

1. **Patient Health Records**
   - Contains patient demographics, vitals, lifestyle indicators, and symptoms.

2. **Hospital Admissions**
   - Contains admission details, treatment information, severity level, and outcomes.

3. **Pharmacy Sales**
   - Contains pharmacy-level medicine sales, prescriptions, and environmental factors.

4. **Doctor Prescriptions**
   - Contains doctor-issued prescriptions linked to diseases and medicines.

5. **Weather Data**
   - Contains environmental conditions such as temperature, humidity, rainfall, and season.

6. **Population Demographics**
   - Contains city-level population statistics and healthcare infrastructure data.

7. **Disease Outbreak Data**
   - Contains reported disease cases, hospitalizations, deaths, and vaccination rates.

8. **Health Tweets**
   - Social media data mentioning diseases for NLP analysis.

9. **Medicine Information**
   - Lookup dataset containing medicine categories, diseases treated, dosage forms, and manufacturers.

## Summary of Findings

After performing the initial data exploration, the following observations were made:

### Dataset Size

The project contains multiple large datasets including:

• 254k patient records  
• 500k hospital admissions  
• 1 million pharmacy sales records  
• 300k prescription records  
• 300k disease outbreak reports  
• 100k health-related tweets  

These datasets provide a rich environment for machine learning, time series analysis, and NLP modeling.

---

### Data Relationships

The datasets are properly linked through key identifiers such as:

• patient_id  
• city  
• medicine_name  
• disease  

These relationships enable multi-source healthcare analytics.

---

### Missing Values

Some datasets contain missing values that will be addressed in the next step:

• Missing alcohol consumption values in patient data  
• Missing disease labels in hospital and prescription datasets  
• Missing side effects in the medicine dataset  

These will be handled during the data cleaning phase.

---

### Data Quality Issues Identified

A few minor issues were detected:

• Negative values in weight and BMI columns  
• Missing disease information in some records  

These issues will be corrected in the next preprocessing step.
