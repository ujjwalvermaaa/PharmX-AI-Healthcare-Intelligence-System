## Encoding and Feature Preparation

In this notebook we convert categorical variables into numerical format so that machine learning models can interpret them.

Machine learning algorithms require numerical inputs, therefore categorical features such as gender, city, disease, and medicine categories must be encoded.

Encoding techniques used in this notebook include:

• Label Encoding  
• One-Hot Encoding  
• Ordinal Encoding  
• Feature Scaling

This notebook prepares all datasets for machine learning and deep learning models.

## Import Libraries

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

## Load Feature Engineered Datasets

```python
patient = pd.read_csv("../data/feature_engineered/patient_features.csv")
hospital = pd.read_csv("../data/feature_engineered/hospital_features.csv")
pharmacy = pd.read_csv("../data/feature_engineered/pharmacy_features.csv")
prescription = pd.read_csv("../data/processed/prescription_clean.csv")
weather = pd.read_csv("../data/processed/weather_clean.csv")
population = pd.read_csv("../data/processed/population_clean.csv")
outbreak = pd.read_csv("../data/feature_engineered/outbreak_features.csv")
tweets = pd.read_csv("../data/feature_engineered/tweets_features.csv")
medicine = pd.read_csv("../data/processed/medicine_clean.csv")
```

### Encoding Patient Dataset

```python
le = LabelEncoder()

patient['gender'] = le.fit_transform(patient['gender'])
patient['smoking_status'] = le.fit_transform(patient['smoking_status'])
patient['alcohol_consumption'] = le.fit_transform(patient['alcohol_consumption'])
patient['physical_activity_level'] = le.fit_transform(patient['physical_activity_level'])
```

```python
patient = pd.get_dummies(patient, columns=['city','region'], drop_first=True)
```

```python
patient['disease'] = le.fit_transform(patient['disease'])
```

```python
import joblib

joblib.dump(le, "../models/disease_encoder.pkl")
```

### Encoding Hospital Dataset 

```python
severity_map = {
'Mild':0,
'Moderate':1,
'Severe':2,
'Critical':3
}

hospital['severity_level'] = hospital['severity_level'].map(severity_map)
```

```python
hospital['disease'] = le.fit_transform(hospital['disease'])
hospital['outcome'] = le.fit_transform(hospital['outcome'])
```

```python
hospital = pd.get_dummies(
    hospital,
    columns=['city','admission_type','treatment_type'],
    drop_first=True
)
```

### Encoding Pharmacy Dataset 

```python
pharmacy['medicine_name'] = le.fit_transform(pharmacy['medicine_name'])
```

```python
pharmacy = pd.get_dummies(
    pharmacy,
    columns=['city','medicine_category'],
    drop_first=True
)
```

### Encoding Prescription Dataset 

```python
prescription['disease'] = le.fit_transform(prescription['disease'])
prescription['medicine_name'] = le.fit_transform(prescription['medicine_name'])
prescription['dosage'] = le.fit_transform(prescription['dosage'])
prescription['frequency'] = le.fit_transform(prescription['frequency'])
prescription['treatment_duration'] = le.fit_transform(prescription['treatment_duration'])
prescription['prescription_type'] = le.fit_transform(prescription['prescription_type'])
```

### Encoding Weather Dataset

```python
weather['season'] = le.fit_transform(weather['season'])
weather = pd.get_dummies(weather, columns=['city'], drop_first=True)
```

### Encoding Population Dataset

```python
population = pd.get_dummies(population, columns=['city','state'], drop_first=True)
```

### Encoding Outbreak Dataset

```python
outbreak['disease'] = le.fit_transform(outbreak['disease'])
outbreak = pd.get_dummies(outbreak, columns=['city'], drop_first=True)
```

### Encoding Tweets Dataset

```python
tweets['sentiment'] = le.fit_transform(tweets['sentiment'])
tweets['disease_mentioned'] = le.fit_transform(tweets['disease_mentioned'])
tweets = pd.get_dummies(tweets, columns=['location'], drop_first=True)
```

### Encoding Medicine Dataset

```python
medicine['medicine_name'] = le.fit_transform(medicine['medicine_name'])
medicine['category'] = le.fit_transform(medicine['category'])
medicine['primary_disease'] = le.fit_transform(medicine['primary_disease'])
medicine['dosage_form'] = le.fit_transform(medicine['dosage_form'])
medicine['dosage_strength'] = le.fit_transform(medicine['dosage_strength'])
medicine['side_effects'] = le.fit_transform(medicine['side_effects'])
medicine['manufacturer'] = le.fit_transform(medicine['manufacturer'])
medicine['prescription_required'] = le.fit_transform(medicine['prescription_required'])
```

## Feature Scaling

Scaling is applied to numerical columns to improve model performance.

```python
scaler = StandardScaler()

numerical_cols = [
'age',
'bmi',
'blood_glucose',
'cholesterol',
'risk_score'
]

patient[numerical_cols] = scaler.fit_transform(patient[numerical_cols])
```

### Save Encoded Datasets

```python
patient.to_csv("../data/encoded/patient_encoded.csv", index=False)
hospital.to_csv("../data/encoded/hospital_encoded.csv", index=False)
pharmacy.to_csv("../data/encoded/pharmacy_encoded.csv", index=False)
prescription.to_csv("../data/encoded/prescription_encoded.csv", index=False)
weather.to_csv("../data/encoded/weather_encoded.csv", index=False)
population.to_csv("../data/encoded/population_encoded.csv", index=False)
outbreak.to_csv("../data/encoded/outbreak_encoded.csv", index=False)
tweets.to_csv("../data/encoded/tweets_encoded.csv", index=False)
medicine.to_csv("../data/encoded/medicine_encoded.csv", index=False)
```

## Encoding Completed

All categorical variables across datasets have been converted into numerical representations using appropriate encoding techniques.

These encoded datasets are now ready for machine learning model training.
