## Feature Engineering

In this notebook we create new predictive features from the cleaned datasets.

Feature engineering transforms raw data into meaningful variables that improve machine learning performance.

The goal is to capture hidden patterns in:

• patient health data  
• hospital admission trends  
• pharmacy medicine demand  
• environmental conditions  
• disease outbreaks  
• social media health signals

These engineered features will be used in machine learning, deep learning, and NLP models.

## Import Libraries

```python
import pandas as pd
import numpy as np
```

## Load Cleaned Data

```python
patient = pd.read_csv("../data/processed/patient_clean.csv")
hospital = pd.read_csv("../data/processed/hospital_clean.csv")
pharmacy = pd.read_csv("../data/processed/pharmacy_clean.csv")
prescription = pd.read_csv("../data/processed/prescription_clean.csv")
weather = pd.read_csv("../data/processed/weather_clean.csv")
population = pd.read_csv("../data/processed/population_clean.csv")
outbreak = pd.read_csv("../data/processed/outbreak_clean.csv")
tweets = pd.read_csv("../data/processed/tweets_clean.csv")
medicine = pd.read_csv("../data/processed/medicine_clean.csv")
```

## Patient Features

### Symptom Count

```python
symptom_cols = [
    'symptom_fever',
    'symptom_cough',
    'symptom_fatigue',
    'symptom_chest_pain',
    'symptom_headache',
    'symptom_shortness_of_breath',
    'symptom_nausea',
    'symptom_joint_pain'
]

patient['symptom_count'] = patient[symptom_cols].sum(axis=1)
```

### Risk Score

```python
patient['risk_score'] = (
    patient['bmi'] * 0.2 +
    patient['blood_glucose'] * 0.3 +
    patient['cholesterol'] * 0.2 +
    patient['blood_pressure_systolic'] * 0.2 +
    patient['age'] * 0.1
)
```

### Lifestyle Risk Score

```python
patient['lifestyle_risk'] = (
    patient['smoking_status'].map({
        'Non-Smoker':0,
        'Former Smoker':1,
        'Smoker':2
    }).fillna(0)
)
```

## Hospital Features

### Admission Duration

```python
hospital['admission_date'] = pd.to_datetime(hospital['admission_date'])
hospital['discharge_date'] = pd.to_datetime(hospital['discharge_date'])

hospital['admission_duration'] = (
    hospital['discharge_date'] - hospital['admission_date']
).dt.days
```

### Cost Per Day

```python
hospital['cost_per_day'] = (
    hospital['treatment_cost'] /
    hospital['length_of_stay']
)
```

## Pharmacy Features

### Medicine Demand Index

```python
pharmacy['demand_index'] = (
    pharmacy['units_sold'] *
    pharmacy['prescriptions_count']
)
```

## Weather Features

### Weather Risk Index

```python
weather['weather_risk_index'] = (
    weather['temperature'] * 0.3 +
    weather['humidity'] * 0.3 +
    weather['rainfall'] * 0.2 +
    weather['air_quality_index'] * 0.2
)
```

## Population Features

### Healthcare Capacity

```python
population['beds_per_1000'] = (
    population['hospital_beds'] /
    population['population']
) * 1000
```

## Outbreak Features

### Fatality Rate

```python
outbreak['fatality_rate'] = (
    outbreak['deaths'] /
    outbreak['cases_reported']
)
```

### Hospitalization Rate

```python
outbreak['hospitalization_rate'] = (
    outbreak['hospitalizations'] /
    outbreak['cases_reported']
)
```

## NLP Prep Features

### Tweet Length

```python
tweets['tweet_length'] = tweets['tweet_text'].apply(len)
```

### Engagement Score

```python
tweets['engagement_score'] = (
    tweets['likes'] + tweets['retweets']
)
```

## Medicine Feature

### Price Category

```python
medicine['price_category'] = pd.qcut(
    medicine['average_price'],
    3,
    labels=['Low','Medium','High']
)
```

## Save Feature Engineered Data

```python
patient.to_csv("../data/feature_engineered/patient_features.csv",index=False)
hospital.to_csv("../data/feature_engineered/hospital_features.csv",index=False)
pharmacy.to_csv("../data/feature_engineered/pharmacy_features.csv",index=False)
outbreak.to_csv("../data/feature_engineered/outbreak_features.csv",index=False)
tweets.to_csv("../data/feature_engineered/tweets_features.csv",index=False)
```
